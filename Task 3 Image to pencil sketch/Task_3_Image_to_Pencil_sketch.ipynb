###Importing Modules

```python
import cv2
from google.colab.patches import cv2_imshow
```

Reading Original image

```python
img = cv2.imread('dog.jpeg')
```

Reducing the size and Displaying Original image

```python
img = cv2.resize(img,None, fx=0.3, fy = 0.3)
cv2_imshow(img)
```

Converting from BGR to Grayscale image

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

Converting to Inverse Grayscale image

```python
inv_img = cv2.bitwise_not(gray)
cv2_imshow(inv_img)
```

Adding Blur to the Inverse Grayscale image

```python
blurred_inv_img = cv2.GaussianBlur(inv_img, (21, 21), 0)
cv2_imshow(blurred_inv_img)
```

Converting to Blurred Grayscale image

```python
blurred_gray_img = cv2.bitwise_not(blurred_inv_img)
cv2_imshow(blurred_gray_img)
```

Final Output: Pencil Sketch

```python
pencil_sketch = cv2.divide(gray, blurred_gray_img, scale=230.0)
cv2_imshow(pencil_sketch)
```
